```python
# Mostrar las primeras filas del dataframe
print('''ejercicio 1 mostrar las primeras filas del dataframe''')
print(df.head(10))
```

```python
# Calcular ventas por region
print('''ejercicio 2 calcular ventas por region''')
# primero va el dataframe
# segundo va la funcion
# tercero va la separacion de la region en este caso
# las llaves me dan el valor numerico
# le doy la funcion sum para que me de el valor total por region
ventas_por_region = df.groupby('Region')['Ventas'].sum()
print(ventas_por_region)
print('\n')
print(df.head(10))
```

```python
# Ejercicio 3: filtrar el DataFrame para mostrar solo las ventas realizadas por david
print('''Ejercicio 3: filtrar el DataFrame para mostrar solo las ventas realizadas por david''')
ventas_david = df[df['Vendedor'] == 'David']
print(ventas_david)
print('\n')
```

```python
# Ejercicio 4: calcular el promedio de ventas por producto
print('''Ejercicio 4: calcular el promedio de ventas por producto''')
ventas_por_producto = df.groupby('Producto')['Ventas'].mean()
print(ventas_por_producto)
print('\n')
```

```python
# Ejercicio 5: Ordenar el DataFrame por ventas en orden descendente
print('''Ejercicio 5: ordenar el DataFrame por ventas en orden decendente''')
df_ordenado = df.sort_values(by='Ventas', ascending=False)
print(df_ordenado)
print('\n')
```

MANIPULACION DE DATOS
- FILTRA LAS FILAS DONDE LAS VENTAS SEAN MAYORES A 10000
- ORDENA EL DATAFRAME EN ORDEN DESCENDENTE
- ELIMINAR LA COLUMNA VENDEDOR

```python
# Ejercicio 6: filtrar las filas donde las ventas sean mayores a 10000
print('''Ejercicio 6: filtrar las filas donde las ventas sean mayores a 10000''')
df_filtrado = df[df['Ventas'] > 10000]
print(df_filtrado)
```

```python
# Ejercicio 7: ordena el dataframe en orden descendente
print('''Ejercicio 7: ordena el dataframe en orden descendente''')
df_ordenado = df.sort_values(by='Ventas', ascending=True)
print(df_ordenado)
```

```python
#Ejercicio 8: eliminar la columna vendedor
print('''Ejercicio 4: elimina la columna vendedor''')
df_sin_vendedor= df.drop(columns=['Vendedor'])
print(df_sin_vendedor)
print('\n')
```

AGREGACION Y APLICACION
- AGRUPAR LAS VENTAS POR REGION
- CALCULAR EL TOTAL DE VENTAS POR PRODUCTO

```python
# Ejercicio 9: agrupar las ventas por region
print('''Ejercicio 9: agrupar las ventas por region''')
ventas_por_region = df.groupby('Region')['Ventas'].sum()
print(ventas_por_region)
```

```python
# Ejercicio 10: calcular el total de ventas por producto
print('''Ejercicio 10: calcular el total de ventas por producto''')
ventas_por_producto = df.groupby('Producto')['Ventas'].sum()
print(ventas_por_producto)
```

CREAR UN DATAFRAME DEL JAMBOARD

```python
#Crear un dataframe de pandas con datos de una persona
data= {
    'id':[1, 2, 3, 4, 5, 3, 5, 6],
    'weight':[144.5, 167.2, 124.1, 144.5, 133.2, 124.1, 129.2, 'none'],
    'height':[5.9, 5.4, 5.2, 5.9, 5.7, 5.2, 5.3, 'none'],
    'age': [33, 45, 23, 33, 54, 23, 42, 'none'],
    'gender': ['M', 'M', 'F', 'M', 'F', 'F', 'M', 'none']
}

df_persona = pd.DataFrame(data)
```

```python
df_persona.head(9)
```

```python
#Datos nulos
df_persona.dropna(inplace=True)
df_persona.info()
```

```python
#Filas repetidas
print(f'tamaño del set antes de eliminar filas repetidas: {df_persona.shape}')
df_persona.drop_duplicates(inplace=True)
print(f'tamaño del set despues de eliminar filas repetidas: {df_persona.shape}')
```
